4. Interpolace a aproximace funkce jedné proměnné

```python
from scipy import interpolate
import numpy as np
from scipy.interpolate import lagrange
import math
import matplotlib.pyplot as plt


#1. metoda - obyčejná metoda
def obycejna_metoda(start,end,posun,x,fx):
    dx = (end - start)/posun
    print(dx)       
    print(x)
            
    plt.title("Interpolace")
    plt.plot(x, fx, "ro")
    plt.show()

#2. metoda - lagrangeova interpolace
def Lagrangeova_metoda(x,fx):
    phi = lagrange(x, fx)

    print(phi)
    print(phi.coef)

    n = 10
    xphi = np.linspace(x[0], x[-1], n)
    plt.title("Lagrangeova interpolace")
    plt.plot(xphi, phi(xphi), "go-")
    plt.plot(x, fx, "ro")
    plt.show()

#3. metoda - lineární interpolace
def Linearni_interpolace(start,konec,posun,x,fx):
    phi = interpolate.interp1d(x, fx)           
    print(phi.x, phi.y)
    xnew = np.arange(start,konec,posun)                 
    ynew = phi(xnew)              

    plt.title("Lineární interpolace")  
    plt.plot(x, fx, 'ro', xnew, ynew, 'gx')     
    plt.show()

print("První funkce")
def f(x):                               
    return x**2 - 4*x + 1
x = np.linspace(0, 10, 5)
fx = f(x)
obycejna_metoda(0,10,2,x,fx)
Lagrangeova_metoda(x,fx)
Linearni_interpolace(0,10,0.2,x,fx)

print("Druhá funkce")
x = np.array([-1, 0, 1, 3])
fx = np.array([2, 1, 2, 0])

obycejna_metoda(0,10,2,x,fx)
Lagrangeova_metoda(x,fx)
Linearni_interpolace(-1,3,0.2,x,fx)

x = np.array([1, 2, 0, 4])
fx = np.array([2, 1, 2, 0])

obycejna_metoda(0,10,2,x,fx)
Lagrangeova_metoda(x,fx)
Linearni_interpolace(0,4,0.2,x,fx)
```
