4. Interpolace a aproximace funkce jedné proměnné

```python
from scipy import interpolate
import numpy as np
import matplotlib.pyplot as plt
#1. metoda - obyčejná metoda
def f(x):                               
    return x**2 - 4*x + 1

start = 0
end = 10
posun = 5
dx = (end - start)/posun
print(dx)
x = np.linspace(start, end, posun)       
print(x)

fx = f(x)                               

plt.title("Interpolace")
plt.plot(x, fx, "ro")
plt.show()

#2. metoda - lagrangeova interpolace
from scipy.interpolate import lagrange

x = np.array([-1, 0, 1, 3])
fx = np.array([2, 1, 2, 0])

phi = lagrange(x, fx)

print(phi)
print(phi.coef)

n = 10
xphi = np.linspace(x[0], x[-1], n)
plt.title("Lagrangeova interpolace")
plt.plot(xphi, phi(xphi), "go-")
plt.plot(x, fx, "ro")
plt.show()

#3. metoda - lineární interpolace
phi = interpolate.interp1d(x, fx)           
print(phi.x, phi.y)
xnew = np.arange(-1, 3, 0.1)                 
ynew = phi(xnew)              

plt.title("Lineární interpolace")  
plt.plot(x, fx, 'ro', xnew, ynew, 'gx')     
plt.show()
```
